```python
#Load the dataset
df = pd.read_csv("amazon.csv")
#Display the first five rows
df.head()
```

```python
#Shape of the dataset
print("Dataset Shape:",df.shape)
#Columns names
print("\nColumn names:")
print(df.columns)
#information about the datset
print("\nDataset innformation:")
df.info()
```

```python
df["discounted_price"] = df["discounted_price"].str.replace("₹", "", regex=False)
df["discounted_price"] = df["discounted_price"].str.replace(",", "", regex=False)
df["discounted_price"] = df["discounted_price"].astype(float)
df["actual_price"]= df["actual_price"].str.replace("₹" ,"" ,regex=False)
df["actual_price"] = df["actual_price"].str.replace(",", "", regex=False)
df["actual_price"] = df["actual_price"].astype(float)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["rating_count"] = df["rating_count"].str.replace(",", "", regex=False)
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")


```

```python
df.dtypes
```

```python
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "", regex=False)

df["discount_percentage"] = pd.to_numeric(df["discount_percentage"], errors="coerce")                                                                  
```

```python
df["rating_count"].fillna(df["rating_count"].median(),inplace=True)
```

```python
import matplotlib.pyplot as plt
df["rating"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Rating")
plt.ylabel("Number of products")
plt.title("Top 10 Ratings")
plt.show()
```

```python
df["category"].value_counts().head(10).plot(kind="bar", figsize=(12,6))
plt.title("top 10 categories")
plt.xlabel("category")
plt.ylabel("Number of products")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["rating"], bins=20)
plt.title("Rating distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
top_reviews = df.sort_values(by="rating_count",ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.bar(top_reviews["product_name"],
top_reviews["rating_count"])
plt.xticks(rotation=90)
plt.title("top 10 most revieed products")
plt.xlabel("product")
plt.ylabel("rating_count")
plt.show()
```

```python
avg_discount = df.groupby("category")["discount_percentage"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(avg_discount.index, avg_discount.values)
plt.title("Top 10 Categories by Average Discount")
plt.xlabel("Category")
plt.ylabel("Average Discount (%)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["discounted_price"], bins=20)
plt.title("Distribution of Discounted Prices")
plt.xlabel("Discounted Price")
plt.ylabel("Number of Products")
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["discounted_price"], df["rating"])
plt.title("Price vs Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[["discounted_price","actual_price","discount_percentage","rating","rating_count"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
